```python
import os
import numpy as numpy
import pandas as pd 
import re
import numpy as np
```

```python
path = os.getcwd()
```

```python
file_list = os.listdir()
keys = {}
```

```python
for files in file_list:
    if files.endswith(".csv"): 
        match = re.search(r'data_(\d+)_',files)
        if match:
            keys[int(match.group(1))] = files
    
```

```python
sorted_keys = sorted(keys.keys())
df_list = []
```

```python
for ks in sorted_keys:
    file = keys[ks]
    df = pd.read_csv(file)
    df_list.append(df)


```

```python
full_df = pd.concat(df_list,ignore_index=True)
```

```python
combined_file = full_df.to_csv('Lead_school_data.csv',index=False)
```

```python
df_copied = df.copy()
df_copied.columns = df_copied.columns.str.replace('\xa0', ' ', regex=False)
col_to_drop = ['School Street','Cleaned_school_name','PS_No','Location','Remediation Status',
               'Type of Organization','Compliance Period','School Website','Date Survey Updated','BEDS Code',
               'County Location','All Results Received','Out of Service or Addressed',
               'School State','Number of Outlets that Require Sampling','Number of Outlets Sampled 2023', 
               'Number of Outlets Sampled 2024','Number of Outlets Sampled 2025', 'Sampling Complete','School ZIP Code','Free_lunch','Total_students']
df_copied.drop(columns=col_to_drop,inplace=True)
df_copied = df_copied.rename(columns={'Number of Outlets, Result ≤ 5 ppb' : 'num_lte_5ppb', 'Number of Outlets, Result > 5 ppb': 'num_gt_5ppb'})
df_copied.columns = df_copied.columns.str.replace(' ', '_').str.strip()
df_copied.columns
```

### Comment

Some of the entries in 'Number of Outlets, Result ≤ 5 ppb' are negative. So, we set those rows to be NaN and also set the corresponding entries in 'Number of Outlets, Result > 5 ppb' to be  NaN

```python
mask = df_copied['num_lte_5ppb'] < 0
df_copied.loc[mask,['num_lte_5ppb','num_gt_5ppb']]= np.nan
```

```python
df_copied.describe()
```

```python
## encodes classification: 1 <--> there exists a lead outlet > 5ppb
df_copied['target_col'] = (df_copied['num_gt_5ppb']>0).astype(int)
```

```python
# clean up 
categorical_cols = ['School_District', 'School_City', 'County', 'School']
for col in categorical_cols: 
    df_copied[col] = df_copied[col].str.lower().str.replace(' ', '_').str.strip()
df_copied
```

### Add feature for % Houses Built Before 1950, by county

```python
df_housing = pd.read_excel("lead.xlsx", sheet_name="Housing") # non-nyc counties
df_nyc_housing = pd.read_csv("NYC EH Data Portal - Very old homes (filtered).csv") #nyc counties
```

```python
# clean up dataframes: column names + drop unnecessary columns
df_housing = df_housing.rename(columns={'Percent\nHousing Built\nBefore 1950': 'percent_housing_pre_1950'})
df_housing['County'] = df_housing['County'].str.lower().str.replace(' ', '_').str.strip()

df_nyc_housing = df_nyc_housing.rename(columns={'Borough': 'County', 'Percent': 'percent_housing_pre_1950'})
df_nyc_housing['County'] = df_nyc_housing['County'].str.lower().str.replace(' ', '_').str.strip()
df_nyc_housing = df_nyc_housing.drop(columns=['TimePeriod', 'GeoTypeDesc', 'GeoID', 'GeoRank', 'BoroID','Geography', 'Area', 'Number' ])
```

```python
# clean up county names
mask = df_copied.County.isin(['man', 'manhattan', 'new_york'])
df_copied.loc[mask, 'County'] = 'manhattan'
len(df_copied.County.unique())
```

```python
set(df_copied.County.unique()) - set(df_housing.County.unique())
```

These are the 5 NYC counties plus St. Lawrence county. Checking, we see that St. Lawrence is recoreded as 'St.Lawrence' in df and 'St. Lawrence' in df_lead. 

```python
SL_spellings = ['st._lawrence', 'st.lawrence', 'st_lawrence']
mask_SL1 = df_copied.County.isin(SL_spellings)
mask_SL2 = df_housing.County.isin(SL_spellings)

df_copied.loc[mask_SL1, 'County'] = "st_lawrence"
df_housing.loc[mask_SL2, 'County'] = "st_lawrence"
```

```python
# check

(set(df_copied.County.unique()) - set(df_housing.County.unique())) or (set(df_housing.County.unique()) - set(df_copied.County.unique()) )
```

Now we have df with all 62 NY state counties, and df_housing with all non-NYC counties. 

```python
ny_housing = pd.concat([df_housing, df_nyc_housing])
```

```python
df_merged = pd.merge(
    df_copied,
    ny_housing,
    on='County',        # both have the same column name
    how='left'          # keeps all rows from df, matches by county to ny_housing df
)
```

```python
# Save as a csv
df_merged.to_csv("lead_data_expanded.csv", index=False)
```
